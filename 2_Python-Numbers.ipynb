```python
print("Python Numbers, Type Conversion and Mathematics\n")


print('''Python supports integers, floating-point numbers and complex numbers. 
They are defined as int, float, and complex classes in Python.\n''')

print("----- Int ----")
#int
Int_vel = 2

print(Int_vel)
print(type(Int_vel))

print("\n ----- Float ----")

#float
Float_vel = 77.0908

print(Float_vel)
print(type(Float_vel))


print("\n ----- complex ----")
#complex
complex_vel = 9 + 5j

print(complex_vel)
print(type(complex_vel))

```

```python
# Python program to convert decimal into other number systems

dec = 344

print("\n ----- decimal ----")
print("The decimal value of", dec, "is:")

print("\n ----- binary ----")
print(bin(dec), "in binary.")

print("\n ----- octal ----")
print(oct(dec), "in octal.")

print("\n ----- hexadecimal ----")
print(hex(dec), "in hexadecimal.\n")

print(bin(568))

# Output: 107
print(0b1101011)

# Output: 253 (251 + 2)
print(0xFB + 0b10)

# Output: 13
print(0o15)
```

```python
#Type Conversion

print('We can convert one type of number into another. This is also known as coercion.\n ')

#Operations like addition, subtraction coerce integer to float implicitly (automatically), if one of the operands is float

print(1 + 2.0)
print(9.0 + 2)

print("\n ----- To INT    ----")
print(int(2.3))

print("\n ----- TO INT    ----")
print(int(-2.8))

print("\n ----- TO FLOAT   ----")
print(float(5))

print("\n ----- TO COMPLEX ----")
print(complex('3+5j'))
```

```python
#Python Decimal

print((1.1 + 2.2) == 3.3)


print("\nWhat is going on?")

print(""" \nIt turns out that floating-point numbers are implemented in computer hardware as 
binary fractions as the computer only understands binary (0 and 1). Due to this reason, 
most of the decimal fractions we know, cannot be accurately stored in our computer.\n""")

print( "1.1 + 2.2 = ", 1.1 + 2.2)


'''To overcome this issue, we can use the decimal module that comes with Python. While floating-point numbers have 
precision up to 15 decimal places, the decimal module has user-settable precision.'''


import decimal

print('\n ------------------Decimal--------------')

print(0.1)
print(decimal.Decimal(0.1))

print('\n ------------------ More Decimal examples --------------')

from decimal import Decimal as D

print(D('1.1') + D('2.2'))

print(D('1.2') * D('2.50'))

```

```python
#Python Fractions

print('\n ---- Python Fractions ----')

'''Python provides operations involving fractional numbers through its fractions module.

A fraction has a numerator and a denominator, 
both of which are integers. This module has support for rational number arithmetic.'''



import fractions

print(fractions.Fraction(1.5))
print(fractions.Fraction(5))
print(fractions.Fraction(1,3))



print('\n ---- As float ----')
# As float
print(fractions.Fraction(1.1))

print('\n ---- As string ----')
# As string
print(fractions.Fraction('1.1'))
```

```python
#Python Mathematics

print('\n ---- Python Mathematics ----') 


print('''\nPython offers modules like math and random to 
carry out different mathematics like trigonometry, logarithms,
probability and statistics, etc.\n ''')

import math

print('\n ---- Examples ----') 
print("\nmath.pi = " , math.pi)
print("\nmath.cos(math.pi)=",math.cos(math.pi))
print("\nmath.exp(10)=",math.exp(10))
print("\nmath.log10(1000)=",math.log10(1000))
print("\nmath.sinh(1)=",math.sinh(1))
print("\nmath.factorial(6)=",math.factorial(6))



```

```python
print('\n ---- Python random ----') 

import random

print('\n ---- Examples ----') 


print(random.randrange(10, 20))

x = ['a', 'b', 'c', 'd', 'e']

print('\n ---- Get random choice ----') 
# Get random choice
print(random.choice(x))

print('\n ---- Shuffle x ----') 
# Shuffle x
random.shuffle(x)

print('\n ---- Print the shuffled x ----') 
# Print the shuffled x
print(x)

print('\n ---- Print random element ----') 
# Print random element
print(random.random())
```
